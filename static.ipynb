```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('year_data.csv')
df['increase'] = df['num'].pct_change()*100
df
```

```python
bar = alt.Chart(df).mark_bar().encode(
    x='year:O',
    y='num:Q',
).properties(
    width=600,
    height=400
)

line = alt.Chart(df).mark_line(color='red').encode(
    x='year:O',
    y='increase:Q'
)

(bar + line).resolve_scale(y='independent').properties(title='Number of ADA Title III Federal Lawsuits Filed per Year')
```
